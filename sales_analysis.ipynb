# Import Necessary Librairies

```python
import pandas as pd
import os
```

## Task #1 Merge the 12 months of sales into a single CSV file

### Get all files in directory

```python
df = pd.read_csv('SalesAnalysis/Sales_Data/Sales_April_2019.csv')

files = [file for file in os.listdir('SalesAnalysis/Sales_Data')]

for file in files:
    print(file)
```

### Concatenate all files

```python
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('SalesAnalysis/Sales_Data/' + file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv('all_data.csv', index=False)
```

### Read in updated dataframe

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

### Clean up the data

#### Drop rows of NaN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

#### Find 'Or' and delete it

```python
temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
temp_df.head()
```

#### Drop rows of Or

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

#### Convert columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # Make float
```

## Task #2 Add Month Column

### Augment data with additional columns

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

## Task #3 Add Sales Column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Q1: What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results['Sales']) # (x_axis, y_axis)
plt.xticks(months)
plt.ylabel('Sales in USD(M$)')
plt.xlabel('Month number')
plt.show()
```

#### Q2: What city had the highest number of sales?

## Task #4 Add City Column

```python
# Use of .apply()

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2][1:3]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + '(' + get_state(x) + ')')
all_data.head()
```

```python
results = all_data.groupby('City').sum()

results
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in all_data.groupby('City')] # making the same order as results['Sales']

plt.bar(cities, results['Sales']) # (x_axis, y_axis)
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD(M$)')
plt.xlabel('City name')
plt.show()
```
